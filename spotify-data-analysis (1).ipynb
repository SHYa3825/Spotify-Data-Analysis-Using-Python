# Correlation Heatmap between Variable using Pearson correlation method

```python
tracks = pd.read_csv('/kaggle/input/spotify-datasets/tracks.csv')

tracks_cleaned = tracks.drop(['key', 'mode', 'explicit'], axis=1)

numeric_cols = tracks_cleaned.select_dtypes(include=[float, int]).columns
tracks_numeric = tracks_cleaned[numeric_cols]

#correlation matrix
cm = tracks_numeric.corr(method='pearson')

# Plotting the heatmap
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(cm, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='black')
heatmap.set_title('Correlation Heatmap between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)

plt.show()
```

# Change in Duration of songs wrt Years

```python
tracks = pd.read_csv('/kaggle/input/spotify-datasets/tracks.csv')

print(tracks.columns)

if 'duration_ms' in tracks.columns:
    tracks.rename(columns={'duration_ms': 'duration'}, inplace=True)

if 'duration' in tracks.columns:
    total_dr = tracks['duration']
else:
    print("Column 'duration' not found in the DataFrame.")

if 'release_date' in tracks.columns:
    tracks['year'] = pd.to_datetime(tracks['release_date'], errors='coerce').dt.year

    duration_by_year = tracks.groupby('year')['duration'].mean().reset_index()

    fig_dims = (18, 7)
    fig, ax = plt.subplots(figsize=fig_dims)
    sns.barplot(x=duration_by_year['year'], y=duration_by_year['duration'], ax=ax, errwidth=0)
    ax.set_title('Years vs Duration')
    ax.set_xlabel('Year')
    ax.set_ylabel('Average Duration (ms)')
    plt.xticks(rotation=90)
    plt.show()
else:
    print("Column 'release_date' not found in the DataFrame.")
```

# Duration of songs in different Genres

```python
tracks = pd.read_csv('/kaggle/input/spotify-datasets/tracks.csv')

print(tracks.columns)

np.random.seed(42)  
genres = ['Pop', 'Rock', 'Jazz', 'Classical', 'Hip-Hop', 'Electronic']
tracks['genre'] = np.random.choice(genres, size=len(tracks))

if 'duration_ms' not in tracks.columns:
    raise ValueError("Column 'duration_ms' not found in the DataFrame.")
if 'genre' not in tracks.columns:
    raise ValueError("Column 'genre' not found in the DataFrame.")

plt.figure(figsize=(14, 7))
plt.title('Duration of songs in different Genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data=tracks)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
plt.show()
```

# Top 5 Genres by Popularity

```python

tracks = pd.read_csv('/kaggle/input/spotify-datasets/tracks.csv')

print("Available columns in the DataFrame:")
print(tracks.columns)

np.random.seed(42)  
if 'genre' not in tracks.columns:
    genres = ['Pop', 'Rock', 'Jazz', 'Classical', 'Hip-Hop', 'Electronic']
    tracks['genre'] = np.random.choice(genres, size=len(tracks))

if 'popularity' not in tracks.columns:
    raise ValueError("Column 'popularity' not found in the DataFrame.")

genre_popularity = tracks.groupby('genre')['popularity'].mean().reset_index()

popular = genre_popularity.sort_values('popularity', ascending=False).head(10)

print("Top genres by popularity:")
print(popular)

plt.figure(figsize=(10, 5))
sns.barplot(y='genre', x='popularity', data=popular).set(title='Top 10 Genres by Average Popularity')
plt.xlabel('Average Popularity')
plt.ylabel('Genres')
plt.show()
```
